```python
# Import the libraries
import os
import numpy as np
import csv
import pandas as pd
import multiprocessing as mp
import shutil
import pyDOE2
```

```python
# This controls the number of samples to take
N = 100

# Generate a random initial point distributed with uniform distribution
points = np.random.uniform(low=-5.0,high=5.0,size=(N,1))


best_other_combination:np.ndarray = np.tile(np.array([-4.941756651838729, -4.954927562502093, 
                                                      -0.512086103576246, -4.896501886781205]),reps=(N,1))
```

```python
# Paste the two arguments
init_points:np.ndarray   = np.hstack((best_other_combination,points),dtype=float)
```

```python
init_points
```

```python
sigma_array = np.array([0.2,1,2]).reshape((3,))
```

```python
sigma_array_comp:np.ndarray = np.vstack((np.tile(sigma_array[0],reps=(N,1)),
                                        np.tile(sigma_array[1],reps=(N,1)),
                                        np.tile(sigma_array[2],reps=(N,1))))

problem_type = np.tile(np.array([2]),reps=(sigma_array_comp.shape[0],1))
lamda_ = np.tile(np.array([2512]),reps=(sigma_array_comp.shape[0],1))

# Final Array
fin_arr:np.ndarray = np.hstack((problem_type,lamda_,sigma_array_comp,np.tile(init_points,reps=(3,1))))

```

```python
# Convert everything to a Pandas array
df:pd.DataFrame = pd.DataFrame(data=fin_arr,columns=['Problem_Type',
                                                     'Penalty_factor',
                                                     'Initial_Step_Size',
                                                     'x0',
                                                     'x1',
                                                     'x2',
                                                     'x3',
                                                     'x4'])
```
